```python
def displayChart(a1, a2):
    birth_chart_svg = KerykeionChartSVG(a1, chart_type="Composite", second_obj=a2)
    birth_chart_svg.makeSVG()
    display(SVG(f"{birth_chart_svg.output_directory}" + '/'+ f"{birth_chart_svg.name}{birth_chart_svg.chart_type}Chart.svg"))
```

```python
def displayTransitChart(a1, day):
    birth_chart_svg = KerykeionChartSVG(a1, chart_type="Transit", second_obj=day)
    birth_chart_svg.makeSVG()
    display(SVG(f"{birth_chart_svg.output_directory}" + '/'+ f"{birth_chart_svg.name}{birth_chart_svg.chart_type}Chart.svg"))
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Создаем граф
G = nx.DiGraph()

# Добавляем связи между планетами
edges = [
    ("Sun in Gem", "Mercury in Gem"),
    ("Moon in Sco", "Mars in Leo"),
    ("Mercury in Gem", "Mercury in Gem"),
    ("Venus in Tau", "Venus in Tau"),
    ("Mars in Leo", "Sun in Gem"),
    ("Jupiter in Gem", "Mercury in Gem"),
    ("Saturn in Ari", "Mars in Leo"),
    ("Uranus in Tau", "Venus in Tau"),
    ("Neptune in Ari", "Mars in Leo"),
    ("Pluto in Aqu", "Saturn in Ari"),
    ("Mean_Node in Pis", "Jupiter in Gem"),
]

# Добавляем ребра в граф
G.add_edges_from(edges)

# Настраиваем визуализацию
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)  # Позиции узлов

# Рисуем граф
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=10, font_weight='bold', arrows=True)

# Сохраняем граф в формате SVG
plt.title("Planetary Relationships")
plt.savefig("planetary_relationships.svg", format="svg")
plt.show()
```
